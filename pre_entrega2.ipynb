**Optimización de Rutas Marítimas para la Marina Mercante utilizando IA**

**Resumen**

Este proyecto tiene como objetivo desarrollar un sistema de optimización de rutas marítimas basado en datos meteorológicos, tráfico marítimo y restricciones portuarias. Se utilizarán herramientas de inteligencia artificial para procesar datos en tiempo real y generar recomendaciones textuales de rutas y visualizaciones de mapas interactivos. La propuesta busca mejorar la eficiencia, reducir costos y aumentar la seguridad operativa para la marina mercante uruguaya.

1. Introducción

La planificación de rutas marítimas eficientes es uno de los desafíos clave para las empresas del sector marítimo. Factores como condiciones meteorológicas cambiantes, tráfico marítimo y restricciones portuarias pueden afectar significativamente los tiempos de navegación, incrementar los costos operativos y aumentar los riesgos para la seguridad.
El presente proyecto propone el desarrollo de un sistema de inteligencia artificial que procese información en tiempo real y genere recomendaciones de rutas óptimas, acompañadas de visualizaciones para la toma de decisiones.

2. Objetivos

Objetivo general:

Optimizar las rutas marítimas de la marina mercante utilizando herramientas de inteligencia artificial basadas en datos meteorológicos, tráfico marítimo y restricciones portuarias.

Objetivos específicos:

Generar recomendaciones de rutas mediante modelos de IA de texto-texto.
Crear mapas visuales de rutas seguras utilizando IA de texto-imagen.
Reducir el tiempo y los costos asociados a la planificación manual de rutas.
Mejorar la seguridad operativa al evitar zonas de riesgo.

3. Metodología

3.1 Recopilación de Datos

Se utilizarán datos meteorológicos, de tráfico marítimo y de restricciones portuarias obtenidos de fuentes confiables, como APIs meteorológicas y sitios oficiales de autoridades portuarias.

3.2 Procesamiento de Datos y Generación de Prompts

Texto-texto: Utilización de IA para generar recomendaciones detalladas de rutas.
Ejemplo de prompt:

"Considerando las siguientes condiciones meteorológicas y restricciones portuarias, genera una recomendación para una ruta desde Montevideo a Buenos Aires."

Texto-imagen: Generación de mapas visuales que identifiquen rutas seguras y zonas de riesgo.

**Ejemplo de prompt:**

"Crea un mapa que muestre las rutas seguras entre Montevideo y Buenos Aires, destacando zonas de alto riesgo por condiciones meteorológicas adversas."

3.3 Visualización de Resultados

Se mostrarán las rutas recomendadas y las zonas de riesgo a través de gráficos y mapas interactivos.

5. Implementación

El sistema se implementará en Jupyter Notebook. Se dividirá en tres etapas:

Recolección de datos meteorológicos y de tráfico marítimo mediante scraping.

6. Resultados

Texto-texto: Generación de recomendaciones claras y útiles para la planificación de rutas.

Texto-imagen: Creación de mapas visuales que destacan zonas de riesgo y rutas seguras, facilitando la toma de decisiones operativas.

```python
# Importar las librerías necesarias
import openai
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Importar las librerías necesarias
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt

# Scraping para obtener datos meteorológicos simulados
url = "https://www.example.com/weather/montevideo"  # URL simulada
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")

# Extraer datos simulados de temperatura y velocidad del viento
temperatura = "20°C"  # Valor simulado
viento = "20 nudos"  # Valor simulado

# Mostrar los datos extraídos
print(f"Temperatura: {temperatura}")
print(f"Velocidad del viento: {viento}")

# Guardar los datos en un DataFrame
data = {"Ubicación": ["Montevideo"], "Temperatura": [temperatura], "Velocidad del viento": [viento]}
df = pd.DataFrame(data)

# Exportar a un archivo CSV para su posterior uso
df.to_csv("datos_meteorologicos.csv", index=False)
print("Datos guardados en datos_meteorologicos.csv")

# Simulación de respuesta texto-texto
print("\nRespuesta texto-texto:")
respuesta_texto = """
Para el trayecto de Montevideo a Buenos Aires, se recomienda seguir la ruta principal hacia el norte,
evitando las aguas abiertas cercanas a la costa de Mar del Plata debido a mareas altas y vientos fuertes.
Se estima un tiempo de navegación de 30 horas, dado el viento de 20 nudos.
Asegúrese de no llegar a Buenos Aires entre las 10:00 y las 16:00 para evitar las restricciones del puerto.
"""
print(respuesta_texto)

# Generar un gráfico de texto-imagen simulado
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title("Mapa de rutas recomendadas (simulación)", fontsize=14)
ax.plot([0, 10], [0, 10], label="Ruta recomendada", linestyle='--', color="green")
ax.fill_between([4, 6], 4, 6, color="red", alpha=0.5, label="Zona de viento fuerte")
ax.fill_between([7, 8], 7, 8, color="blue", alpha=0.5, label="Zona de mareas altas")
ax.set_xlabel("Coordenadas X")
ax.set_ylabel("Coordenadas Y")
ax.legend()
plt.show()
```
